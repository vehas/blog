```python
def count_same_consicutive(keys):
    old_k = 0
    longest_count = 0
    li = 0
    consicutive_count = 0
    for i, k in enumerate(keys):
        if np.array_equal(old_k, k):
            consicutive_count = consicutive_count + 1
        else:
            if consicutive_count > longest_count:
                longest_count = consicutive_count
                li = i
            consicutive_count = 0
        old_k = k
    return longest_count, li
```

```python
# count_unique(np_img[100])
vv = []
for ni in np_img:
    # c = [1000000*n1 + 1000*n2 + n3 for n1, n2, n3 in ni]
    cu = count_same_consicutive(ni)
    # v = sorted(zip(cu[1], cu[0]), reverse=True)[:3]
    vv.append(cu)
# vv
vv
```
